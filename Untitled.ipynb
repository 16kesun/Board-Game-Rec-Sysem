```python
#get rec based on game

def sim_rec(game, k=5):
    """
    This function takes in a name of a game and gives you k recommendations for other games that you may like
    """
    iid = basic.trainset.to_inner_iid(game) #changes the name of the beer to its id
    
    recs = basic.get_neighbors(iid, k) #takes the id and gives you k recommendations for it (also in id form)
    
    for rec in recs:
        print(basic.trainset.to_raw_iid(rec)) #takes the id and converts that back into a beer name and prints it
```

```python
#create function that removes brackets from categories

categories_list = []
for i in range(len(game_info.boardgamecategory)):
    if pd.isnull(game_info.boardgamecategory[i]) == True:
        x = game_info.boardgamecategory[i]
        continue
    
    elif pd.isnull(game_info.boardgamecategory[i]) == False:
        x = game_info.boardgamecategory[i]
        x = x.replace('[','')
        x = x.replace(']','')
        x = x.replace("'","")
        x = x.split(",")
        categories_list.append(x)

big_list = []
for lst in categories_list:
    for i in lst:
        big_list.append(i)

unique_list = []
for i in big_list:
    if i not in unique_list:
        unique_list.append(i)
    else: continue
len(unique_list)

def unique(column):
    
    """Returns list of unique values for the
    column that was inputed"""
    
    categories_list = []
    for i in range(len(column)):
        if pd.isnull(column[i]) == True:
            x = column[i]
            continue
        elif pd.isnull(column[i]) == False:
            x = column[i]
            x = x.replace('[','')
            x = x.replace(']','')
            x = x.replace("'","")
            x = x.split(",")
            categories_list.append(x)

    big_list = []
    for lst in categories_list:
        for i in lst:
            big_list.append(i)

    unique_list = []
    for i in big_list:
        if i not in unique_list:
            unique_list.append(i)
        else: continue
    return unique_list
```
